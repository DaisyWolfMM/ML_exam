```python
import numpy as np
import scipy as sp
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import numpy.random as rnd
rnd.seed(42)#1337

from collections import namedtuple

from operator import itemgetter
from itertools import accumulate, chain
from functools import reduce
from matplotlib import pyplot
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
pred =pd.DataFrame(columns = [0,1])
```

```python
Data= pd.read_csv('test_data.csv', sep = ',', encoding = 'cp1251')
for i in range(1,201):#1,201
    Data= pd.read_csv('test_data.csv', sep = ',', encoding = 'cp1251')
    series = Data.T[i-1]
    series=series[1:]
    #для предсказание каждой точки
    history = [x for x in series]
    #только предсказание
    predictions = list()
    #для одного ряда
    for t in range(1,101):
        model = ARIMA(history, order=(7,1,0))
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(float(yhat))
        obs = yhat
        history.append(obs)

    pred=pred.append(pd.DataFrame([[str(i)+'_'+str(j) for j in range(1,101)],list(predictions)]).T)
    print(i)
```

```python
pred.columns = ['Id','Predicted']
```

```python
pred.to_csv('kaggle_prediction_arima_7.csv', index=False)
```
